```python
def change(n):
    '''
    Arg:
        n: 거슬러줘야 할 돈
    Return:
        coin_n: 거슬러줘야 할 동전의 최소 개수
    '''
    currency = [500,100,50,10]
    coin_n = 0
    for a in currency:
        coin_n += n // a
        n %= a
    return coin_n
```

### 예제 3-2 큰 수의 법칙

```python
def big_nums(nums, data):
    '''
    주어진 수를 m번 더해서 가장 큰 수를 만드는 법칙
    Args:
        nums:
            n: 배열의 크기
            m: 숫자가 더해지는 횟수
            k: 큰 수를 더할 수 있는 최대 횟수
        data:
            실제 데이터
    Output: 
        큰 수의 법칙에 따라 더해진 답
    '''
    _, m, k = map(int, nums.split())
    data = sorted(list(map(int, data.split())))
    
    # 가장 큰 수가 더해지는 횟수 계산
    cnt = int(m/(k+1)) * k
    cnt += m % (k+1)

    return cnt * data[-1] + (m-cnt) * data[-2] 
    

    
```

```python
big_nums('5 8 3', '2 4 5 4 6')
```

### 예제 3-3 숫자 카드 게임

```python
def num_cards(size, cards):
    '''
    Args:
        size: 행의 개수 n, 열의 개수 m이 공백을 기준으로 하는 자연수로 주어짐
        cards: n개의 줄에 걸쳐 각 카드에 대한 숫자가 주어짐
    Return: 게임의 룰에 맞게 선택한 카드에 적힌 숫자
    
    룰: 여러 개의 숫자 카드 중 가장 높은 숫자가 쓰인 카드 한 장을 뽑도록 함
        행 선택 -> 그 행에서 가장 숫자가 낮은 카드를 뽑음
    '''
    n, _ = map(int, size.split())
    cards = [list(map(int, cards.splitlines()[i].split())) for i in range(n)]
    _min = [min(line) for line in cards]
    rst = max(_min)

    return rst

num_cards('3 3', '3 1 2\n4 1 4\n2 2 2')
```

### 예제 4-1 상하좌우
- 인풋: 공간의 크기 N // 명령어 (L, R, U, D)
- 아웃풋: 최족 도착할 지점의 좌표 (X, Y)를 공백으로 구분해 출력

공간의 크기를 벗어나는 명령어는 무시

```python
def move(N, s):
    c = s.split()  # 주어진 명령 리스트 생성
    c_space = [1, 1]  # 초기 위치 설정

    def command(c_, current_space):
        command_list = {'U': [0, -1], 'D': [0, 1], 'L': [-1, 0], 'R': [1, 0]}
        return [current_space[i] + command_list[c_][i] for i in range(2)]

    for i in range(N):
        rst = command(c[i], c_space)
        if 1 <= rst[0] <= N and 1 <= rst[1] <= N:
            print(f'{rst}로 이동 성공!')
            c_space = rst
        else:
            print(f'{rst}로 이동 불가')

    return ' '.join([str(place) for place in c_space])


N = 5
s = "R R R U D D"
print(move(N, s))  # Expected output: "3 2"
```

### 실전 문제 3. 게임 개발


```python
def game_developing(size, c_state, info):
    '''
    Args:
        size: 맵의 세로 크기 n과 가로 크기 m을 공백으로 구분해 입력
        c_state: 게임 캐릭터가 있는 칸의 좌표(A, B)와 바라보는 방향 d가 서로 공백으로 구분해 주어짐
        info: n개의 줄에 맵의 상태가 북쪽부터 남쪽 순서대로, 각 줄의 데이터는 서쪽부터 동쪽 순서대로 주어짐. 
              맵의 외곽은 항상 바다. (0은 육지, 1은 바다) / 처음에 게임 캐릭ㅌ터가 위치한 칸의 상태는 항상 육지
    Return:
        이동을 마친 후 캐릭터가 방문한 칸의 수
    '''
    # 맵 생성
    n, m = map(int, size.split())
    d = [[0]*m for _ in range(n)]

    #현재 캐릭터 좌표, 방향
    x, y, _dir = map(int, c_state.split())
    d[x][y] = 1 #현재 위치 방문 처리

    #전체 맵 정보
    array = [list(map(int, i.split())) for i in info.splitlines()]

    #방향 정의 (북, 동, 남, 서)
    dx = [-1,0,1,0]
    dy = [0,1,0,-1]

    #왼쪽으로 회전
    def turn_left(direction):
        direction -= 1
        if direction == -1:
            direction = 3
        return direction

    cnt = 1
    turn_time = 0
    
    while True:
        #왼쪽으로 회전
        _dir = turn_left(_dir)
        nx = x + dx[_dir]
        ny = y + dy[_dir]
        #회전한 뒤 정면에 안 가본 칸이 있으면
        if d[nx][ny] == 0 and array[nx][ny] == 0:
            d[nx][ny] = 1
            x = nx
            y = ny
            cnt += 1
            turn_time = 0
        else:
            turn_time += 1
        if turn_time == 4:
            nx = x - dx[_dir]
            ny = y - dy[_dir]
            if array[nx][ny] == 0:
                x = nx
                y = ny
            else:
                break
            turn_time = 0
    
    return cnt
```

# DFS / BFS

- stack: append, pop
- queue: append, popleft
- 리스트 대신 from collections import deque, queue에 append, pop으로 더 효율적인 메모리 관리!

### DFS
1. **스택** 자료 구조 이용
2. 동작 과정
    1. 탐색 시작 노드를 스택에 삽입하고 방문 처리 (다시 해당 노드가 삽입되지 않게 체크)
    2. 스택의 최상단 노드에 방문하지 않은 인접 노드가 있으면 그 인접 노드를 스택에 넣고 방문 처리. 
        
        방문하지 않은 인접 노드가 없으면 스택에서 최상단 노드 꺼냄
    3. 2의 과정을 더이상 수행할 수 없을 때까지 반복

### BFS
1. **큐** 자료 구조 이용
2. 동작 과정   
    1. 탐색 시작 노드를 스택에 삽입하고 방문 처리
    2. 큐에서 노드를 꺼내 해당 노드의 인접 노드 중 방문하지 않은 노드를 모두 큐에 삽임한 뒤 방문 처리
    3. 2의 과정을 더이상 수행할 수 없을 때까지 반복

```python
# 큐 예제

from collections import deque

queue = deque()

queue.append(5)
queue.append(2)
queue.append(3)
queue.append(7)
queue.popleft()
queue.append(1)
queue.append(4)
queue.popleft()

print(queue)
queue.reverse()
print(queue)
```

```python
# DFS 예제

def dfs(graph, v, visited):
    visited[v] = True
    print(f'{v}에 방문했습니다.')

    #현재 노드와 연결된 다른 노드들을 재귀적으로 방문
    for i in graph[v]:
        if not visited[i]:
            dfs(graph, i, visited)
```

```python
graph = [
    [],
    [2,3,8],
    [1,7],
    [1,4,5],
    [3,5],
    [3,4],
    [7],
    [2,6,8],
    [1,7]
]

#각 노드가 방문된 정보를 리스트 자료형으로 표현
visited = [0]*9

dfs(graph, 1, visited)
```

```python
# BFS 예제

from collections import deque

def bfs(graph, start, visited):
    queue = deque([start])
    visited[start] = True
    while queue:
        v = queue.popleft()
        print(f'{v}를 방문했습니다.')
        for i in graph[v]:
            if not visited[i]:
                queue.append(i)
                visited[i] = True
                
                
```

### 실전 문제 5-3 음료수 얼려 먹기
1. 특정 지점 주변 상하좌우를 본 뒤, 주변 지점 중 값이 0이면서 아직 방문하지 않은 지점이 있으면 해당 지점 방문
2. 방문한 지점에서 다시 상하좌우를 보면서 방문을 재진행하며 연결된 모든 지점 방문
3. 위의 과정을 모든 노드에 반복하며 방문하지 않은 지점의 수 세기

```python
def drink(size, arr):
    n, m = map(int, size.split())
    graph = [list(map(int, line.split())) for line in arr.strip().splitlines()]
    
    def dfs(x, y):
        # 주어진 범위를 벗어나면 종료
        if x < 0 or x >= n or y < 0 or y >= m:
            return False
        if graph[x][y] == 0:  # 아직 방문 안 했으면
            graph[x][y] = 1
            dfs(x - 1, y)
            dfs(x, y - 1)
            dfs(x + 1, y)
            dfs(x, y + 1)
            return True
        return False

    rst = 0
    for i in range(n):
        for j in range(m):
            if dfs(i, j) == True:
                rst += 1

    return rst
```

# 6. 정렬

정렬 알고리즘으로 정렬하면 이진 탐색이 가능해진다. 종류로는 선택 정렬, 삽입 정렬, 퀵 정렬, 계수 정렬, 그 외 다른 방식들이 있으며 파이썬에서 제공하는 기본 정렬 라이브러리도 있다. 

## 1. 선택 정렬: 가장 작은 것을 선택 $O(N^2)$
가장 작은 데이터를 선택해 맨 앞에 있는 데이터와 바꾸고, 그 다음 작은 데이터를 선택해 앞에서 두 번째 데이터와 바꾸는 과정을 반복하는 정렬. 

```python
array = [7,5,9,0,3,1,6,2,4,8]
for i in range(len(array)):
    min_idx = i
    for j in range(i+1, len(array)):
        if array[min_idx] > array[j]:
            min_idx = j
    array[i], array[min_idx] = array[min_idx], array[i] #swap

print(array)
```

## 2. 삽입 정렬: 특정한 데이터를 적절한 위치에 삽입 $O(N^2)$
필요할 때만 위치를 바꿔, 데이터가 거의 정렬되어 있을 때 후러씬 효율적 
- 선택 정렬은 현 데이터 상태와 무관하게 무조건 모든 원소를 비교하고 위치를 바꿈

```python
array = [7,5,9,0,3,1,6,2,4,8]
for i in range(1, len(array)):
    for j in range(i,0,-1): #i부터 1까지 감소하며 반복하는 문법
        if array[j] < array[j-1]: #한 칸씩 왼쪽으로 이동
            array[j], array[j-1] = array[j-1], array[j]
        else:
            break #자기보다 작은 데이터를 만나면 그 자리에서 멈춤
print(array)
```

## 3. 퀵 정렬: 기준 데이터를 설정하고 그 기준과 비교 == pivot을 사용 $O(N\log N)$
*병합 정렬 알고리즘도 빠름

- 호어 분할 (Hoare Partition) 방식으로 퀵 정렬 설명하기
    1. 리스트에서 첫 번째 데이터를 피벗으로 설정
    2. 왼쪽에서 피벗보다 큰 데이터, 오른쪽에서 피벗보다 작은 데이터 찾으며 피벗에 대한 정렬 수행

```python
array = [5,7,9,0,3,1,6,2,4,8]

def quick_sort(array, start, end):
    if start >= end:
        return #원소가 한 개일 경우 종료
    pivot = start #피벗은 첫 번째 원소
    left = start + 1
    right = end
    while left <= right:
        #피벗보다 큰 데이터 찾을 때까지 반복
        while left <= end and array[left] <= array[pivot]:
            left += 1
        #피벗보다 작은 데이터 찾을 때까지 반복
        while right > start and array[right] >= array[pivot]:
            right -= 1
        #왼쪽이 오른쪽보다 커졌을 경우 피벗을 작은 데이터와 교체
        if left > right:
            array[right], array[pivot] = array[pivot], array[right]
        #왼쪽이 오른쪽보다 작을 경우 작은 데이터와 큰 데이터를 교체
        else:
            array[left], array[right] = array[right], array[left]
    #분할 후 왼쪽, 오른쪽 나눠 각각 정렬 수행
    quick_sort(array, start, right-1)
    quick_sort(array, right+1, end)

quick_sort(array,0, len(array)-1)
print(array)
```

```python
# 파이썬 식으로 고친다면
def py_quick_sort(array):
    if len(array) <= 1:
        return array
    pivot = array[0]
    tail = array[1:]

    left = [x for x in tail if x <= pivot]
    right = [x for x in tail if x > pivot]

    return py_quick_sort(left) + [pivot] + py_quick_sort(right)

py_quick_sort(array)
```

# 7. 이진 탐색

## 1. 순차 탐색: 리스트 안에 있는 특정한 데이터를 찾기 위해 앞에서부터 차례대로 찾는 방식
## 2. 이진 탐색: 반으로 쪼개면서 탐색하기 $O(\log N)$
배열 내부의 데이터가 정렬되어 있어야만 사용할 수 있음.

탐색하고자 하는 범위의 시작점, 끝점, 중간점을 사용해, 찾으려는 데이터와 중간점 위치에 있는 데이터를 반복적으로 비교해 원하는 데이터를 찾는 방법

한 단계를 거칠 때마다 확인하는 원소가 평균적으로 절반으로 줄어듦. 

재귀함수를 이용하는 방법, 반복문을 이용하는 방법 두 가지

```python
def binary_search(array, target, start, end): #재귀함수로 구현하면
    if start > end:
        return None
    mid = (start + end) // 2
    #찾은 경우 중간점 인덱스 반환
    if array[mid] == target:
        return mid
    #중간점 값보다 타겟이 작은 경우 왼쪽 확인
    elif array[mid] > target:
        return binary_search(array, target, start, mid-1)
    else:
        return binary_search(array, target, mid + 1, end)

def binary_search(array, target, start, end):
    while start <= end:
        mid = (start+end) // 2
        if array[mid] == target:
            return mid
        elif array[mid] > target:
            end = mid - 1
        else:
            start = mid + 1
    return None
```

```python
d = [0]*100

#top-down dynamic programming
def fibo(x):
    if x == 1 or x == 2:
        return 1
    if d[x] != 0:
        return d[x]
    d[x] = fibo(x-1) + fibo(x-1)
    return d[x]

print(fibo(99))

#bottom-up
d = [0]*100
d[1], d[2] = 1,1
n = 99
for i in range(3, n+1):
    d[i] = d[i-1] + d[i-2]

print(d[n])
```

```python
#다익스트라 알고리즘 (간단 버전)
from sys import stdin
input = sys.stdin.readline
INF = int(1e9)

n,m = map(int, input().split())
start = int(input())
graph = [[] for _ in range(n+1)]
visited = [0]*(n+1)
distance = [INF]*(n+1)

for _ in range(m):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))

def get_smallest_node():
    min_value = INF
    index = 0
    for i in range(1, n+1):
        if distance[i] < min_value and not visited[i]:
            min_value = distance[i]
            index = i
    return index

def dijkstra(start):
    distance[start] = 0
    visited[start] = 1
    for j in graph[start]:
        distance[j[0]] = j[1]
    for i in range(n-1):
        now = get_smallest_node() #현재 최단거리가 가장 짧은 노드를 꺼내 방문 처리
        visited[now] = 1
        #현재 노드와 연결된 다른 노드 확인
        for j in graph[now]:
            cost = distance[now] + j[1]
            if cost < distance[j[0]]:
                distance[j[0]] = cost #현재 노드를 거쳐서 다른 노드로 이동하는 거리가 더 짧을 경우

dijkstra(start)

#모든 노드로 가기 위한 최단 거리를 출력
for i in range(1, n+1):
    #도달할 수 없으면 INFINITY라고 출력
    if distance[i] == INF:
        print('INFINITY')
    else:
        print(distance[i])
```

```python
#개선된 다익스트라 알고리즘
#최단 거리가 가장 짧은 노드를 빠르게 찾자!
import heapq
from sys import stdin
input = stdin.readline
INF = int(1e9)

n,m = map(int, input().split())
start = int(input())
graph = [[] for _ in range(n+1)]
visited = [0]*(n+1)
distance = [INF]*(n+1)

for _ in range(m):
    a, b, c = map(int, input().split())
    graph[a].append((b, c))

def dijkstra(start):
    q = []
    heapq.heappush(q, (0, start))
    distance[start] = 0
    while q: #큐가 비어 있지 않다면
        dist, now = heapq.heappop(q)
        if distance[now] < dist:
            continue
        for i in graph[now]:
            cost = dist + i[1]
            if cost < distance[i[0]]:
                distance[i[0]] = cost
                heapq.heappush(q, (cost, i[0]))

dijkstra(start)

#모든 노드로 가기 위한 최단 거리를 출력
for i in range(1, n+1):
    #도달할 수 없으면 INFINITY라고 출력
    if distance[i] == INF:
        print('INFINITY')
    else:
        print(distance[i])
```

구조대 : 119

전화번호부에 적힌 전화번호를 담은 배열 phone_book 이 solution 함수의 매개변수로 주어질 때, 어떤 번호가 다른 번호의 접두어인 경우가 있으면 false를 그렇지 않으면 true를 return 하도록 solution 함수를 작성해주세요.



```python
def solution(phone_book):
    for i in range(len(phone_book)):
        for j in range(i, len(phone_book)):
            if phone_book[i] in phone_book[j] or phone_book[j] in phone_book[i]:
                return True
            
    return False
```
