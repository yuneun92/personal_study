```python
# 1. 0~9까지 중 없는 숫자 더하기

def solution(numbers):
    return sum([i for i in range(10) if i not in numbers])

# 이런 방법이!
def solution(numbers):
    return 45 - sum(numbers)
# 이런 방법도!
solution = lambda x: sum(range(10)) - sum(x)

# 2. () 스택

def solution(s):
    count = 0
    for string in s:
        if string == '(':
            count += 1
        elif string == ')':
            count -= 1
        if count <= -1:
            return False
    if count != 0:
        return False
    return True

# 이것도 알아두자
def is_pair(s):
    st = list()
    for c in s:
        if c == '(':
            st.append(c)

        if c == ')':
            try:
                st.pop()
            except IndexError:
                return False

    return len(st) == 0

# 3. 피보나치
def fib(num):
    if num == 0 or num == 1:
        return num
    elif num == 2:
        return 1
    
    a,b = 1,1
    for i in range(1,num): # 이 부분 기억하기
        a,b = b, a+b

    return a

def solution(n):
    return fib(n) % 1234567

# 4. 제곱근 여부 확인

def solution(n):
    sqrt_n = n ** 0.5
    if int(sqrt_n) ** 2 == n:
        return (sqrt_n+1)**2
    else:
        return -1

# 5. p와 y의 개수

def solution(s):
    s = s.lower()
    if s.count('p') == s.count('y'):
        return True
    return False


# 6. 정수 내림차순으로 배치하기


def solution(n):
    answer = 0
    num = sorted([int(s) for s in list(str(n))])
    for i in range(len(num)):
        answer += num[i]*(10**i)
    return answer

# 문자열로 접근하면
def solution(n):
    ls = list(str(n))
    ls.sort(reverse = True)
    return int("".join(ls))


# 7. 수박 무한 반복

def solution(n):
    return '수박'*(n//2) + '수'*(n%2)


# 8. 문자열 중간값

def solution(s):
    id = len(s)//2
    if len(s)%2 == 0:
        return s[id-1:id+1]
    return s[id]
#다른 풀이
def string_middle(str):
    return str[(len(str)-1)//2 : len(str)//2 + 1]


# 9. 카카오톡 다음달 선물 많이 받을 친구 구하기

# 처음 내 코드인데, value값 다 잘 나오는데도 자꾸 오류 남 (디버깅 필요)
def solution(friends, gifts):
    # next: 다음 달에 받을 선물 수, score: 선물 지수
    wish_dict = {k:0 for k in friends}
    score_dict = {k:0 for k in friends}
    # 선물 주고 받은 기록을 딕셔너리로 관리
    gift_dict = {k:{j:0 for j in friends if k != j} for k in friends}

    for log in gifts:
        k, v = log.split()
        score_dict[k] += 1 # 선물 지수 조정
        score_dict[v] -= 1
        if k not in gift_dict.keys():
            gift_dict[k] = {v:1}
        else:
            if v not in gift_dict[k].keys():
                gift_dict[k][v] = 1
            else:
                gift_dict[k][v] += 1
    # 다음 달 선물을 누가 많이 받을까요?   
    for a in friends:
        for b in friends:
            if a != b:
                if gift_dict[a][b] != 0: #a가 b에게 준 적이 있고
                    if not gift_dict[b][a] or gift_dict[a][b] > gift_dict[b][a]: #b는 a에게 준 적이 없거나 더 적게 줬으면
                        wish_dict[a] += 1
                        print(f'{b}는 다음달에 {a}에게 선물을 줍니다')
                    elif gift_dict[a][b] < gift_dict[b][a]: #b가 선물 더 많이 줬으면
                        wish_dict[b] += 1
                        print(f'{a}는 다음달에 {b}에게 선물을 줍니다')
                    else: # 주고 받은 개수가 같으면
                        if score_dict[a] > score_dict[b]:
                            wish_dict[a] += 1
                            print(f'{b}는 다음달에 {a}에게 선물을 줍니다')
                        elif score_dict[a] < score_dict[b]:
                            wish_dict[b] += 1
                            print(f'{a}는 다음달에 {b}에게 선물을 줍니다')
    
    print(f'다음 달에 받을 선물 수, 선물 지수를 보여줄게요. {wish_dict}')
    print(f'선물을 주고 받은 기록이에요. {gift_dict}')
    
    return max(wish_dict.values())

# 다른 사람 풀이 1
# 하나의 함수로 하지 말고 항목 별로 나눠서 생각하기 (합성함수)

from collections import defaultdict #그냥 딕셔너리 말고 디폴트 딕셔너리 써서 오류 방지

def solution(friends, gifts):
    answer = 0
    # 선물을 주고받은 기록
    history = {
        friend: {"gift": defaultdict(int), "received": defaultdict(int), "rank": 0}
        for friend in friends
    }
    set_friends = set(friends)

    # 선물 지수
    def history_index(friend):
        x = sum(history[friend]["gift"].values())
        y = sum(history[friend]["received"].values())
        return x - y

    # 선물을 주고받지 않은 사람
    def history_difference(friend):
        y = set(
            list(history[friend]["gift"]) + list(history[friend]["received"]) + [friend]
        )
        return set_friends - y

    # 선물 기록 갱신
    for gift in gifts:
        giver, recipient = gift.split()
        history[giver]["gift"][recipient] += 1
        history[recipient]["received"][giver] += 1

    # 받은 선물 계산
    for friend in friends:
        pivot_index = history_index(friend)
        pivot_difference = history_difference(friend)

        # 선물을 주고받은 기록이 하나도 없을때 처리
        if pivot_difference:
            for name in pivot_difference:
                if pivot_index > history_index(name):
                    history[friend]["rank"] += 1

        # 선물을 주고받은 기록이 있을때 처리
        for giver, give_cnt in history[friend]["gift"].items():
            if giver in history[friend]["received"]:
                received_cnt = history[friend]["received"][giver]
                if give_cnt > received_cnt or (
                    give_cnt == received_cnt and pivot_index > history_index(giver)
                ):
                    history[friend]["rank"] += 1
            else:
                history[friend]["rank"] += 1

        # 가장 많이 선물받은 값 저장
        answer = max(answer, history[friend]["rank"])

    return answer

# 간결하게!
def solution(friends, gifts):
    f = {v: i for i, v in enumerate(friends)}
    l = len(friends)
    p = [0] * l
    answer = [0] * l
    gr = [[0] * l for i in range(l)]
    for i in gifts:
        a, b = i.split()
        gr[f[a]][f[b]] += 1
    for i in range(l):
        p[i] = sum(gr[i]) - sum([k[i] for k in gr])

    for i in range(l):
        for j in range(l):
            if gr[i][j] > gr[j][i]:
                answer[i] += 1
            elif gr[i][j] == gr[j][i]:
                if p[i] > p[j]:
                    answer[i] += 1
    return max(answer)

# 10. 크기가 작거나 같은 부분 문자열

def solution(t, p):
    n = int(p)
    l = len(p)
    cnt = 0
    for i in range(len(t) - len(p) + 1):
        if int(t[i:i+l]) <= n:
            cnt += 1
            print(f'{int(t[i:i+l])}는 {n}보다 작거나 같아요')
    
    return cnt
```
